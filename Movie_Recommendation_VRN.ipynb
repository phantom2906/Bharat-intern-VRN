```python

# Import libraries
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from fuzzywuzzy import process  # Import process from fuzzywuzzy for fuzzy matching

# Assuming your data files are uploaded to Google Colab and paths are correct
# Load data
user_ratings_df = pd.read_csv("/content/ratings_small.csv")
movie_metadata = pd.read_csv("/content/movies_metadata.csv", on_bad_lines='skip')

# Create user-item matrix
user_item_matrix = user_ratings_df.pivot(index='userId', columns='movieId', values='rating').fillna(0)

# Initialize KNN model
cf_knn_model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10, n_jobs=-1)

# Fit model on matrix
cf_knn_model.fit(user_item_matrix)

# Recommendation function
def movie_recommender_engine(movie_name, matrix, cf_model, movie_names, n_recs):
    # Fit model on matrix (if necessary)
    cf_model.fit(matrix)
    # Example of fuzzy matching if using fuzzywuzzy
    movie_id = process.extractOne(movie_name, movie_names['title'])[2]
    # Calculate nearest neighbors
    distances, indices = cf_model.kneighbors(matrix.iloc[movie_id].values.reshape(1, -1), n_neighbors=n_recs)
    # Generate recommendations
    movie_rec_ids = sorted(list(zip(indices.squeeze().tolist(), distances.squeeze().tolist())), key=lambda x: x[1])[:0:-1]
    # List to store recommendations
    cf_recs = []
    for i in movie_rec_ids:
        cf_recs.append({'Title': movie_names['title'].iloc[i[0]], 'Distance': i[1]})
    # Select top number of recommendations needed
    if len(cf_recs) < n_recs:
        df = pd.DataFrame(cf_recs)
    else:
        df = pd.DataFrame(cf_recs, index=range(1, n_recs + 1))
    return df

# Example usage
n_recs = 10
recommended_movies = movie_recommender_engine('Batman', user_item_matrix, cf_knn_model, movie_metadata, n_recs)
print(recommended_movies)
```
